```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special # for factorial
def gauss(x,mu=0,sig=1): return 1/(np.sqrt(2*pi*sig**2))*np.exp(-(x-mu)**2/(2*sig**2))
```

#  (1) Prior Distribution of $\theta$
$$ P(\theta) = 2 - |\sin(\theta))| $$

```python
pi = np.pi
n_view = 180
theta = np.linspace(-pi,pi-(2*pi/n_view),n_view)
ori = np.arange(n_view)
p_theta = 2 - np.abs(np.sin(theta))
plt.plot(theta,p_theta,'k')
plt.xlabel('$\Theta$')
plt.ylabel('$P(\Theta)$')
plt.title('Prior Distribution')
plt.show()
```

```python
# CDF
plt.subplot(1,3,(1,2))
c_theta = np.cumsum(p_theta)
c_theta/=(c_theta[-1]/(2*pi))
c_theta-=pi
plt.plot(theta,c_theta,'k')
pts = np.linspace(0,len(c_theta)-1,20,dtype=int)
plt.plot(theta[pts],c_theta[pts],'ko')
plt.xlabel('$\Theta$')
plt.ylabel('$D(\Theta)$')
plt.title('Tuning Function Distribution')
# plt.show()
plt.subplot(1,3,3)
plt.hist(c_theta,30,color='k',orientation='horizontal')
plt.tight_layout()
plt.show()
```

# (2) Response of $i^{th}$ neuron (w/ pref $\Psi_i$)

```python
M = 20
gamma = .5
kappa = 5
psi = np.linspace(-pi,pi,M) # neuron selectivity
fi = np.zeros((n_view,M))
for i in range(M):
    fi[:,i] = (gamma/M) * np.exp(kappa*np.cos(theta-psi[i])-1)
    
plt.plot(theta,fi)
plt.xlabel('$\Theta$')
plt.ylabel('Neural Response')
plt.title('Receptive Fields (Uniform-Prior) [$\kappa$=%.1f]' %kappa)
plt.show()
```

# (3) Response of $i^{th}$ neuron (non-uniform prior)

```python
D = np.cumsum(p_theta)
D/=D[-1]
D = D*2*pi-pi

fib = np.zeros((n_view,M))
for i in range(M):
    fib[:,i] = (gamma/M) * np.exp(kappa*np.cos(D-psi[i])-1)
    
plt.plot(theta,fib)
plt.xlabel('$\Theta$')
plt.ylabel('Neural Response')
plt.title('Receptive Fields (Non-uniform-Prior) [$\kappa$=%.1f]' %kappa)
plt.show()
print('This is how each of %d neurons will respond to all stimuli' %M)
```

# (4-5) Devisive Normalization and Poisson Spike Generation
- Get responses for a population of 50 neurons for 5 stimuli
- Use rate to generate possion firing processes
$$ p(n_i|\theta,T,N) = \frac{(r_i(\theta,N)T)^{n_i}}{n_i!} e^{-r_i(\theta,N)T} $$

```python
def fs(stim): 
    out = np.ones((len(stim),M))
    Du = D[stim]
    for i in range(len(stim)):
        foo = np.exp(kappa*np.cos(Du[i]-psi)-1)
        out[i,:]=  foo
    out*=gamma/M
    return out
        
# stim = np.random.randint(0,180,5)
stim = np.linspace(0,170,15,dtype=int)
# stim = [40]
n_rate = fs(stim)


plt.plot(stim,n_rate)
plt.xlabel('Orientation Stimuli')
plt.show()
print('This is how %d neurons will respond to %d stimuli' %(M,len(stim)))

this_rate = 3.5
rate_p = np.random.poisson(this_rate,10000)
bns = np.linspace(-.5,10.5,24)

plt.hist(rate_p,bns)
plt.xlabel('# Spikes')
plt.ylabel('# Events')
plt.title('Poisson Process (%.1f spks/s)' %this_rate)
plt.show()

```

# Tie together generative model
- Keep population of M=12 neurons
- generate spike rates for neurons for 10 stimuli
- So output should be M x trials of spike rates

```python
stim = np.random.randint(0,180,1)
spks=np.random.poisson(fs(stim)) # generated spikes

# poisson expected
n = np.arange(20)
def pos_exp(theta):
    resp_exp=[]
    for ni in n:
        resp_exp.append(np.array(fs(theta)**ni/scipy.special.factorial(ni) * np.exp(-fs(theta))))
    return np.array(resp_exp)
```

# Decode Orientation from generative model
For each neuron, there is a likelihood for each number of spikes and orientation.

The likelihood of a given orientation given the number of spikes is:
 - the product of the likelihood of each neurons activation function
 - corrected by the prior likelihood across all possible orientations

```python
plt.plot(spks[0])
plt.ylabel('Spike Rate')
plt.xlabel('Neuron ID')
plt.show()
```

```python
# calculate likelihood for all possible orientations
exp_out = pos_exp(ori)
ll = np.ones(180)
for i in range(M):
    this_ll = exp_out[spks[0][i],:,i]
    ll*=this_ll
```

```python
plt.plot(ll)
dec_stim = np.argmax(ll)
plt.title('Decode: %d deg' %(dec_stim))

plt.plot(dec_stim,ll[dec_stim],'ko')
plt.plot(stim,ll[stim],'ro')
plt.xlabel('Orientation')
plt.ylabel('Likelihood')
plt.show()
```

# Do encoding and decoding together and with more stimuli
- Switch to log-likelihood
- correct for prior

```python
M = 20
psi = np.linspace(-pi,pi,M) # neuron selectivity
nstim=50000
gamma = .4 # width
kappa = 5  # population gain
l_conv = 1000
# for kappa in [4]:
#     for gamma in [.5]:
sav_name = 'SimOutput%d_M%d_gamma%d_kappa%d.png' %(nstim,M,np.round(gamma*10),kappa)
#try:

stim = np.random.randint(0,180,nstim)
spks=np.random.poisson(fs(stim)) # generated spikes
n_spks = np.sum(spks,1)
exp_out = pos_exp(ori)

# get log-likelihood for each stimuli! 
ll = np.zeros((nstim,180))
for s in range(nstim):
    for i in range(M):
        this_ll = exp_out[spks[s][i],:,i]
        ll[s,:] += np.log(this_ll)
ll = np.exp(ll)
dec_stim = np.argmax(ll,1)

stim = stim[n_spks>0]
dec_stim = dec_stim[n_spks>0]

E = dec_stim-stim
E[np.abs(E)>90]-=180*np.sign(E[np.abs(E)>90])

order = np.argsort(stim)
conv_win = gauss(np.linspace(-5,5,l_conv))
conv_win2 = gauss(np.linspace(-1,1,l_conv))
conv_win/=np.sum(conv_win)
conv_win2/=np.sum(conv_win2)
c_absE = np.convolve(np.abs(E[order]),conv_win,mode='valid')
c_absE2 = np.convolve(np.abs(E[order]),conv_win2,mode='valid')
c_E = np.convolve(E[order],conv_win,mode='valid')
c_E2 = np.convolve(E[order],conv_win2,mode='valid')

pos_stim = np.convolve(stim[order],conv_win,mode='valid')
plt.subplot(121)
plt.plot(pos_stim,c_absE)
plt.plot(pos_stim,c_absE2,'r')
plt.title('Oblique Effect')
plt.ylabel('|Error| (deg)')
plt.xlabel('Orientation Stim (deg)')

plt.subplot(122)
plt.plot(pos_stim,c_E)
plt.plot(pos_stim,c_E2,'r')
#         plt.title('Cardinal Bias')
plt.title('Cardinal Bias ($\gamma$:%.2f $\kappa$:%d)' %(gamma,kappa))
plt.plot([0,180],[0,0],'k--')
plt.ylabel('Bias (deg)')
plt.xlabel('Orientation Stim (deg)')
plt.tight_layout()
plt.savefig(sav_name)
plt.show()
        #except:
        #    print(sav_name,'Failed')
```

# Lets make stand alone function that can run given
- M
- Ntrials
- gamma (gain)
- kappa (width)


```python
import numpy as np
from math import factorial

pi = np.pi
n_view = 180
theta = np.linspace(-pi,pi-(2*pi/n_view),n_view)
ori = np.arange(n_view)
p_theta = 2 - np.abs(np.sin(theta)) # prior distribution

D = np.cumsum(p_theta) # CDF
D/=D[-1]
D = D*2*pi-pi

n_spks_sim = np.arange(20) 

def fs(stim): 
    out = np.ones((len(stim),M))
    Du = D[stim]
    for i in range(len(stim)):
        foo = np.exp(kappa*np.cos(Du[i]-psi)-1)
        out[i,:]=  foo
    out*=gamma/M
    return out

# poisson expected n spikes given rate
def pos_exp(theta):
    resp_exp=[]
    for ni in n_spks_sim:
        resp_exp.append(np.array(fs(theta)**ni/factorial(ni) * np.exp(-fs(theta))))
    return np.array(resp_exp)

def run_TBmodel(N=5000,M=20,kappa=3.8,gamma=0.5):
    
    psi = np.linspace(-pi,pi,M) # neuron selectivity
    stim = np.random.randint(0,180,nstim)
    
    spks=np.random.poisson(fs(stim)) # generated spikes
    exp_out = pos_exp(ori) # prior expectation for # of spikes
    
    
    # get log-likelihood for each stimuli! 
    ll = np.zeros((nstim,180))
    for s in range(nstim):
        for i in range(M):
            this_ll = exp_out[spks[s][i],:,i]
            ll[s,:] += np.log(this_ll)
    ll = np.exp(ll)
    dec_stim = np.argmax(ll,1)
    
    n_spks = np.sum(spks,1)
    
    stim = stim[n_spks>0]
    dec_stim = dec_stim[n_spks>0]
    print('Throwing away %d trials with no spikes' %sum(spks==0))
    
    return stim, dec_stim
    
    
```
